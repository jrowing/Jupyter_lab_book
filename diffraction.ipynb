# Determination of wavelength using a diffraction grating

## Theory
The diffraction grating equation is given by $ n \lambda = s \sin{\theta}$. The spacing between the lines in a diffraction grating is usually specified or can be found from the grating ruling. By measuring the angle $\theta$, the wavelength of the light can be determined.

## Apparatus:

* Laser pen
* Diffraction grating of known d value or ruling e.g. 300 lines cm-1
* Metre rule
* Screen
* Stand and clamp for laser pen and grating
